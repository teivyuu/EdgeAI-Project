```python
import pandas as pd
import plotly.graph_objects as go
import glob
```

```python
file_path_lst = glob.glob('/content/drive/MyDrive/Document/EdgeAI/Data/FallDetection/Data/raw/fall*acc.csv')
i = 0
```

```python
if i < len(file_path_lst):
  file_path = file_path_lst[i]
  df = pd.read_csv(file_path, names=['time', 'total', 'x', 'z', 'y'])
  fig = go.Figure([
      go.Scatter(x = df['time'], y = df['total'], name='Total'),
      go.Scatter(x = df['time'], y = df['x'], name='X'),
      go.Scatter(x = df['time'], y = df['y'], name='Y'),
      go.Scatter(x = df['time'], y = df['z'], name='Z'),
      ])
  fig.update_layout(title='Normal',
                    plot_bgcolor='rgb(230, 230,230)',
                    showlegend=True)
  fig.show()  
  i += 1
else:
  print('done')  
```

```python
min_value_time = 1442 #@param {type:"integer"}
max_value_time = 2100 #@param {type:"integer"}
df = df[df['time'] >= min_value_time]
df = df[df['time'] <= max_value_time]
fig = go.Figure([
    go.Scatter(x = df['time'], y = df['total'], name='Total'),
    go.Scatter(x = df['time'], y = df['x'], name='X'),
    go.Scatter(x = df['time'], y = df['y'], name='Y'),
    go.Scatter(x = df['time'], y = df['z'], name='Z'),
    ])

fig.update_layout(title='Normal',
                   plot_bgcolor='rgb(230, 230,230)',
                   showlegend=True)

fig.show()
```

```python
df.to_csv(file_path.replace('/raw', '/labeled'))
print(file_path.replace('/raw', '/labeled'))
```
